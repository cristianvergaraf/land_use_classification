Ahora usando geopandas y leemos un archivo geopackage de QGIS con los sitios de entrenamiento

```python
sitios_entrenamiento_2021 = gpd.read_file('./sitios_entrenamientos_lingue.gpkg', layer = 'sitios_entranamientos_lingue')
sitios_entrenamiento_2021.crs
```

# Para poder extraer los valores de la imagen necesitamos la coordenadas de cada punto

```python

coords = [(x,y) for x, y in zip(sitios_entrenamiento_2021.geometry.x, sitios_entrenamiento_2021.geometry.y)]
coords
```

Usando list comprension obtenemos los valores para cada banda añadiendo cada lista como una columna en el dataframe de los sitios de entrenamiento

```python
sitios_entrenamiento_2021['B1'] = [x[0] for x in lingue_basin.sample(coords, indexes=1)]
sitios_entrenamiento_2021['B2'] = [x[0] for x in lingue_basin.sample(coords, indexes=2)]
sitios_entrenamiento_2021['B3'] = [x[0] for x in lingue_basin.sample(coords, indexes=3)]
sitios_entrenamiento_2021['B4'] = [x[0] for x in lingue_basin.sample(coords, indexes=4)]
sitios_entrenamiento_2021['B5'] = [x[0] for x in lingue_basin.sample(coords, indexes=5)]
sitios_entrenamiento_2021['B6'] = [x[0] for x in lingue_basin.sample(coords, indexes=6)]
sitios_entrenamiento_2021['B7'] = [x[0] for x in lingue_basin.sample(coords, indexes=7)]
```
